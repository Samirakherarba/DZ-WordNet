```python

def modify_column(df):
    il_verbs = df[df["Mot en fr"].str.startswith("il a")]["Daridja arabe"]

    modified_il_verbs = "ن" + il_verbs

    df.loc[df["Mot en fr"].str.startswith("il a"), "1p"] = modified_il_verbs

    def check_3_letters(word):
        if isinstance(word, str):
            return len(word) == 3 and all('\u0600' <= char <= '\u06FF' for char in word)
        else:
            return False

    three_letter_words = df[df["Daridja arabe"].apply(lambda x: check_3_letters(x))]

    filtered_words = three_letter_words[three_letter_words['Type'] == "verbe"]

    df.loc[df["Daridja arabe"].isin(filtered_words["Daridja arabe"]), "1p"] = "ن" + filtered_words["Daridja arabe"]

    return df

df = modify_column(df)
```
