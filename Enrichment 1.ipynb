```python
import requests
from bs4 import BeautifulSoup


def get_definition(word):
    url = f"https://www.larousse.fr/dictionnaires/francais/{word}"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        definition = soup.find('li', class_='DivisionDefinition')
        if definition:
            return definition.text
        else:
            return "Definition not found"
    else:
        return "request denied"


df['Définitions'] = df['Mot en fr'].apply(get_definition)
def raccourcir_definition(definition):
    first_part = definition.split(':')[0]
    clean_part = first_part.split('(')[0].strip()  
    clean_part = clean_part.split('.\r\n\t\t\t\tSynonyme')[0].strip()
    clean_part = clean_part.split('\r\n\t\t\t\tSynonymes')[0].strip()
    if clean_part.startswith('1.'):
        clean_part = clean_part.replace('1.', '', 1).strip() 
    return clean_part

df['Définitions'] = df['Définitions'].apply(raccourcir_definition)
```

```python


def find_synonyms(french_word, daridja_word, df):
    if pd.notna(french_word) and pd.notna(daridja_word):
        synonyms = df[(df['fr word'].str.lower() == french_word.lower()) & (df['Daridja ar'] != daridja_word)]['Daridja ar'].unique()
        return ', '.join(synonyms) if len(synonyms) > 0 else None
    return None

df['Synonym Daridja'] = df.apply(lambda row: find_synonyms(row['fr word'], row['Daridja ar'], df), axis=1)

```

Add the antonyms

```python
import requests
from bs4 import BeautifulSoup

def get_antonyms(word):
    url = f"https://www.larousse.fr/dictionnaires/francais/{word}"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        antonyms_tags = soup.find_all('p', class_='Synonymes')
        antonyms_set = set()
        for tag in antonyms_tags:
            if tag.find_previous('p', class_='LibelleSynonyme') and tag.find_previous('p', class_='LibelleSynonyme').text.strip() == 'Contraires :':
                antonyms_with_links = [antonym.text.strip() for antonym in tag.find_all('a')]
                antonyms_set.update(antonyms_with_links)
        if antonyms_set:
            return ', '.join(antonyms_set)
        else:
            return "Antonyms not found"
    else:
        return "request denied"

df['Antonyms en fr'] = df['Mot en fr'].apply(get_antonyms)
```

```python
def find_daridja_equivalent(antonyms_en_fr):
    words = str(antonyms_en_fr).split(', ')
    daridja_equivalents = []
    for word in words:
        if df['Mot en fr'].str.contains(word).any():
            daridja_equivalent = df.loc[df['Mot en fr'].str.contains(word), 'Daridja arabe'].values
            if len(daridja_equivalent) > 0:
                daridja_equivalents.extend(map(str, daridja_equivalent))
    return ', '.join(daridja_equivalents)


df['Antonyms Daridja'] = df['Antonyms en fr'].apply(find_daridja_equivalent)
```

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

def get_word_relations(word, english_words):
    relations = []
    synsets = wordnet.synsets(word)
    for synset in synsets:
        # Hypernyms
        for hypernym in synset.hypernyms():
            related_word = hypernym.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "hypernym"))
        # Hyponyms
        for hyponym in synset.hyponyms():
            related_word = hyponym.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "hyponym"))
        # Meronyms
        for meronym in synset.part_meronyms():
            related_word = meronym.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "meronym"))
        for meronym in synset.substance_meronyms():
            related_word = meronym.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "substance_meronym"))
        for meronym in synset.member_meronyms():
            related_word = meronym.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "member_meronym"))
        # Holonyms
        for holonym in synset.part_holonyms():
            related_word = holonym.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "holonym"))
        for holonym in synset.substance_holonyms():
            related_word = holonym.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "substance_holonym"))
        for holonym in synset.member_holonyms():
            related_word = holonym.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "member_holonym"))
        # Entailments
        for entailment in synset.entailments():
            related_word = entailment.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "entailment"))
        # Antonyms
        for lemma in synset.lemmas():
            for antonym in lemma.antonyms():
                related_word = antonym.name().split('.')[0]
                if related_word in english_words:
                    relations.append((related_word, "antonym"))
        # Causes Causality
        for cause in synset.causes():
            related_word = cause.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "cause"))
        # Also see
        for also_see in synset.also_sees():
            related_word = also_see.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "also_see"))
        # Verb groups
        for verb_group in synset.verb_groups():
            related_word = verb_group.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "verb_group"))
        # Attributes Attribute Relations
        for attribute in synset.attributes():
            related_word = attribute.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "attribute"))
        # Similar to Near-Synonymy
        for similar_to in synset.similar_tos():
            related_word = similar_to.name().split('.')[0]
            if related_word in english_words:
                relations.append((related_word, "similar_to"))

    return relations



df["relations"] = df["English"].apply(lambda word: get_word_relations(word, set(df["English"])))

lemmatizer = WordNetLemmatizer()
df['English_normalized'] = df['English'].apply(lemmatizer.lemmatize)

df["relations"] = df["English_normalized"].apply(lambda word: get_word_relations(word, set(df["English_normalized"])))


def create_translation_mapping(df):
    translation_mapping = {}
    for index, row in df.iterrows():
        english_word = row['English']
        arabic_word = row['Daridja arabe']
        if english_word not in translation_mapping:
            translation_mapping[english_word] = {arabic_word}
        else:
            translation_mapping[english_word].add(arabic_word)
    return translation_mapping

translation_mapping = create_translation_mapping(df)

def replace_english_with_arabic(relations):
    new_relations = []
    for relation in relations:
        english_word = relation[0]
        if english_word in translation_mapping:
            arabic_words = translation_mapping[english_word]
            for arabic_word in arabic_words:
                new_relation = (arabic_word, relation[1])
                new_relations.append(new_relation)
        else:
            new_relations.append(relation)
    return new_relations

df['relations'] = df['relations'].apply(replace_english_with_arabic)


def remove_self_relations(relations, word):
    new_relations = []
    for relation in relations:
        if word not in relation[0]:
            new_relations.append(relation)
    return new_relations

def process_relations(row):
    word = row['Daridja arabe'] 
    relations = row['relations']  
    if isinstance(relations, list) and relations:  
        relations = remove_self_relations(relations, word)
    return relations

df['relations'] = df.apply(process_relations, axis=1)
```
